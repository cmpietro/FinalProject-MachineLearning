```python
df = pd.read_csv(os.path.join("Q2_labor_char_v10.csv"))
df.head()
```

```python
df1 = df[['Bin_Five_min_APGAR_score_code', 'Anesthesia', 'Antibiotics_for_Mother','Induction_of_Labor', 'Steroids','Average_Age_Mother_years',
         'Average_Birth_Weight_grams','Average_Prenatal_Visits','Average_OE_Gestational_Age_weeks','Average_Pre-pregnancy_BMI']]
df1.head()
```

```python
one_hot_df1 = pd.get_dummies(df1, prefix=['Anesthesia', 'Antibiotics_for_Mother','Induction_of_Labor', 'Steroids'])
one_hot_df1.head()
```

```python
X = one_hot_df1.iloc[:, 1:3].values
y = one_hot_df1.iloc[:, 0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=27)
```

```python
model = XGBClassifier(n_estimators=1000,colsample_bylevel=0.8,max_depth=6,) 
model.fit(X_train, y_train) 
print(model)
```

```python
y_pred = model.predict(X_test)
predictions =[round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" %(accuracy*100))
```
