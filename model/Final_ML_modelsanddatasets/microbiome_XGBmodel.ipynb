```python
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
from sklearn import model_selection, metrics
from sklearn.model_selection import GridSearchCV
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
df = pd.read_csv(os.path.join("microbiome_dataset_prevalence_bf.csv"))
df.head()
```

```python
df1 = df[['Month', 'Delivery_type', 'Infant_Prevalence','Mother_Prevalence','Genus']]
df1.head()
```

```python
one_hot_df1 = pd.get_dummies(df1, prefix=['Delivery_type','Genus'])
one_hot_df1.head()
```

```python
target = one_hot_df1["Month"]
target_names = ["newborn", "4M"]
```

```python
data = one_hot_df1.drop("Month", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target,  test_size=0.3,random_state=72)
```

```python
model = XGBClassifier(n_estimators=1000,colsample_bylevel=0.8, gamma=10) 
model.fit(X_train, y_train) 
print(model)
```

```python
y_pred = model.predict(X_test)
predictions =[round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" %(accuracy*100))
```
