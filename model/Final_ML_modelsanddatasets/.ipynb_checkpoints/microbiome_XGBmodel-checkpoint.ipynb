```python
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
from sklearn import model_selection, metrics
from sklearn.model_selection import GridSearchCV
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
df = pd.read_csv(os.path.join("microbiome_dataset_prevalence_bf.csv"))
df.head()
```

```python
df1 = df[['Month', 'Delivery_type', 'Infant_Prevalence','Mother_Prevalence','Genus']]
df1.head()
```

```python
one_hot_df1 = pd.get_dummies(df1, prefix=['Delivery_type','Genus'])
one_hot_df1.head()
```

```python
target = one_hot_df1["Month"]
target_names = ["newborn", "4M"]
```

```python
data = one_hot_df1.drop("Month", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target,  test_size=0.3,random_state=72)
```

```python
model = XGBClassifier(n_estimators=100,colsample_bylevel=0.8, gamma=10) 
eval_set = [(X_train, y_train), (X_test,y_test)]
model.fit(X_train, y_train, eval_metric=["error","logloss"], eval_set=eval_set, verbose=True) 
print(model)
```

```python
y_pred = model.predict(X_test)
predictions =[round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" %(accuracy*100))
```

```python
#retrieve performance metrics
results =model.evals_result()
epochs=len(results['validation_0']['error'])
x_axis =range(0,epochs)
```

```python
#plot log loss
fig,ax=plt.subplots()
ax.plot(x_axis, results['validation_0']['logloss'], label='Train')
ax.plot(x_axis, results['validation_1']['logloss'], label='Test')
ax.legend()
plt.ylabel('Log loss')
plt.title('XGBoost Log loss')
plt.show()
```

```python
#plot classification error
fig, ax = plt.subplots()
ax.plot(x_axis, results['validation_0']['error'], label='Train')
ax.plot(x_axis, results['validation_1']['error'], label='Test')
ax.legend()
plt.ylabel('Classification error')
plt.title('XGBoost Classification error-  microbiome analysis')
plt.show()
fig.savefig('microbiome_error.png')
```

```python
#avoid overfitting model
model = XGBClassifier(n_estimators=7,colsample_bylevel=0.8, gamma=10) 

eval_set = [(X_test, y_test)]
model.fit(X_train, y_train, eval_metric="error", eval_set=eval_set, verbose=True)
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
